```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_by_artist.csv")
```

```python
df.columns
```

```python
df = df.drop(columns=['duration_ms','popularity', 'key','count'], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler



minmax = MinMaxScaler()
minmax.fit(df[['loudness', 'tempo']])
df_minmax = minmax.transform(df[['loudness', 'tempo']])
```

```python
df_arranged = pd.DataFrame(zip(df_minmax[:,0], df_minmax[:,1]), columns=['minmax_loudness', 'minmax_tempo'])
```

```python
df = pd.merge(df[['artists', 'acousticness', 'danceability', 'energy',
       'instrumentalness', 'liveness', 'speechiness', 'valence', 'mode']], 
                  df_arranged, 
                  left_index=True, 
                  right_index=True)
```

```python
df
```

```python
X = df[['acousticness', 'danceability' ,'energy',
       'instrumentalness', 'liveness', 'minmax_loudness', 'speechiness', 'minmax_tempo',
       'valence', 'mode']]
y = df['artists']
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X, y)
```

-----------------------------------------------------------------------

Example with User Input ( Whatever artist)

```python
from IPython.core import display as ICD

artist_input = str(input('Give an artist : '))

results = knn.kneighbors(X = df[['acousticness', 'danceability', 'energy',
       'instrumentalness', 'liveness', 'minmax_loudness', 'speechiness', 'minmax_tempo',
       'valence', 'mode']][df['artists'] == artist_input], 
                         n_neighbors=3, 
                         return_distance=True)


print("\n")
print(df.artists[results[1].tolist()[0][1]])

output = df[(df['artists'] == artist_input) |(df['artists'] == df.artists[results[1].tolist()[0][1]])]
output2 = output.T
output2.columns = output2.iloc[0]
output2 = output2.drop('artists', axis=0)
output2['Diff'] = output2[artist_input] - output2[df.artists[results[1].tolist()[0][1]]]
ICD.display(output2)


try: 
    from googlesearch import search 
except ImportError:  
    print("No module named 'google' found") 
  
# to search 
query = df.artists[results[1].tolist()[0][1]]
for j in search(query, tld="co.in", num=3, stop=3, pause=2): 
    print(j)
    
print("\n")

print(df.artists[results[1].tolist()[0][2]])

output3 = df[(df['artists'] == artist_input) |(df['artists'] == df.artists[results[1].tolist()[0][2]])]
output4 = output3.T
output4.columns = output4.iloc[0]
output4 = output4.drop('artists', axis=0)
output4['Diff'] = output4[artist_input] - output4[df.artists[results[1].tolist()[0][2]]]
ICD.display(output4[output4['Diff'] != 0])

try: 
    from googlesearch import search 
except ImportError:  
    print("No module named 'google' found") 
  
# to search 
query = df.artists[results[1].tolist()[0][2]]
for j in search(query, tld="co.in", num=3, stop=3, pause=2): 
    print(j) 
```
